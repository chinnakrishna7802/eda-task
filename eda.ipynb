## CASE1

```python
temp_d=df[(df['Specialization']=='computer science & engineering') | (df['Specialization']=='computer science and technology') | (df['Specialization']=='computer science')]
```

```python
import pandas as pd
from scipy import stats

claim_lower_bound = 250000
claim_upper_bound = 300000

t_stat, p_value = stats.ttest_1samp(temp_d['Salary'], (claim_lower_bound + claim_upper_bound) / 2)

alpha = 0.05

if p_value < alpha:
    print(f"The p-value ({p_value:.4f}) is less than alpha ({alpha}), reject the null hypothesis.")
else:
    print(f"The p-value ({p_value:.4f}) is greater than alpha ({alpha}), fail to reject the null hypothesis.")

print(f"Claimed Salary Range: {claim_lower_bound} to {claim_upper_bound}")
print(f"Sample Mean Salary: {df['Salary'].mean()}")
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df['Specialization'], df['Gender'])

# Perform chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Set the significance level
alpha = 0.05

# Check the p-value against the significance level
if p < alpha:
    print(f"There is a significant relationship between Gender and Specialization p-value: {p}.")
else:
    print(f"There is no significant relationship between Gender and Specialization p-value: {p}.")
```
